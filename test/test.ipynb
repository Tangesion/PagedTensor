```python
import torch
from tqdm import tqdm

# 加载模型权重
model_weight2 = torch.load("../../../models/Llama-2-7b-hf/pytorch_model-00001-of-00002.bin", weights_only=True, map_location='cpu')

# 创建一个新的字典来存储所需的权重
new_model_weight = {}

# 选择所需的权重
keys_to_extract = [
    "model.embed_tokens.weight",
    "model.layers.0.self_attn.q_proj.weight",
    "model.layers.0.self_attn.k_proj.weight",
    "model.layers.0.self_attn.v_proj.weight",
    "model.layers.0.self_attn.o_proj.weight",
    "model.layers.0.self_attn.rotary_emb.inv_freq",
    "model.layers.0.mlp.gate_proj.weight",
    "model.layers.0.mlp.up_proj.weight",
    "model.layers.0.mlp.down_proj.weight",
    "model.layers.0.input_layernorm.weight",
    "model.layers.0.post_attention_layernorm.weight"
]

# 提取所需的权重
for key in keys_to_extract:
    if key in model_weight2:
        new_model_weight[key] = model_weight2[key]
        print(f"Extracted {key}: {model_weight2[key].shape}")

# 将新的权重字典保存到一个新的 .bin 文件中
torch.save(new_model_weight, "test_weights.bin")
print("New model weights saved to test_weights.bin")
```
